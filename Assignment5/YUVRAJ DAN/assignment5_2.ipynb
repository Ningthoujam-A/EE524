# K-means 

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing the dataset
dataset=pd.read_csv("C:\\Users\\DELL\\Desktop\dataset.csv")
X=dataset.iloc[:,[1,2]].values
```

```python
#using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss=[]
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
#plotting
plt.plot(range(2,11),wcss)
plt.title('The elbow method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

For K=4

```python

    kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
    y_kmeans=kmeans.fit_predict(X)
    print(y_kmeans)
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=50,c='red',label='cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=50,c='green',label='cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=50,c='cyan',label='cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=50,c='black',label='cluster 4')
plt.title('clusters of dataset')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.legend()
plt.show()
```
