# Rastrigin function.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import axes3d
%matplotlib notebook

n=input("press (1:: 2D and 2:: 3D)")
n=int(n)


#defining basic elements of functions
a=np.linspace(-5,5,100)
b=np.linspace(-10,10,100)


X,Y=np.meshgrid(a,b)

fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(projection="3d")
ax2=fig.add_subplot(projection="3d")

#function implementation
func_1 = 10+(X**2 - 10 * np.cos(2 * np.pi * X))
func_2= 20+(b**2 - 10 * np.cos(2 * np.pi * X))+(Y**2 - 10 * np.cos(2 * np.pi * Y)) 


if n==1:
    func_1 = 10+(a**2 - 10 * np.cos(2 * np.pi * a))
    plt.plot(a,func_1)
    plt.xlabel("X")
    plt.ylabel("func_1")
    plt.grid()
    
if n==2:
    #Drawing the figure
    ax2=fig.add_subplot(projection="3d")
    ax2.plot_surface(X,Y,func_2,color="c")
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('FUNCTION')
    plt.show()


    


```

# Ackley's function.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import axes3d
%matplotlib notebook


inp=input("press (1::forx_1 only or 2::x1,x2 both")
n=int(inp)
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection="3d")
#defining basic elements of function
x_1=np.linspace(-5,5,100)
x_2=np.linspace(-10,10,100)

x_1,x_2=np.meshgrid(x_1,x_2)
x_1_sqr=x_1**2
x_2_sqr=x_2**2
    
#function implementation

func_1=-20*np.exp(-.2*(x_1_sqr)**.5)-np.exp(np.cos(2*np.pi*x_1))+20
func_2=-20*np.exp(-.2*(.5*(x_1_sqr+x_2_sqr))**.5)-np.exp(.5*(np.cos(2*np.pi*x_1)+np.cos(2*np.pi*x_2)))+20
if n==2: 
    #plotting
    ax.plot_surface(x_1,x_2,func_2,cmap=cm.nipy_spectral)
    
    #setting scales
    ax.set_xlabel('x_1')
    ax.set_ylabel('y_2')
    ax.set_zlabel('FUNC')
if n==1:
    #ploting
    ax.plot_surface(x_1,x_2,func_1,cmap=cm.nipy_spectral)
    
    #setting scales
    ax.set_xlabel('x_1')
    ax.set_ylabel('y_2')
    ax.set_zlabel('FUNC')
    

    



plt.show()
```

## SPHERE FUNCTION

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import axes3d
%matplotlib notebook

#defining the basic elements of the function
x_1=np.linspace(-5,5,100)
x_2=np.linspace(-10,10,100)

x_1,x_2=np.meshgrid(x_1,x_2)
x_1_sqr=x_1**2
x_2_sqr=x_2**2

#function implementation
func_1=x_1_sqr
func_2=x_2_sqr+func_1

#drawing figure
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection="3d")
ax.plot_surface(x_1,x_2,func_2,cmap=cm.nipy_spectral)


#setting scales
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('func_2')

plt.show()
```

# Rosenbrock function.

# McCormick function.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import axes3d
%matplotlib notebook


#defining the basic elements of the function
x=np.linspace(-5,5,100)
y=np.linspace(-10,10,100)
x,y=np.meshgrid(x,y)


#function implementation
func=np.sin(x+y)+(x-y)**2-1.5*x+2.5*y+1

#setting figure environment and drawing
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection="3d")
ax.plot_surface(x,y,func,cmap=cm.nipy_spectral)


#setting scales
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('FUNC')


plt.show()
```

# Matyas function.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import axes3d
%matplotlib notebook





#defining the basic elements of the function
x=np.linspace(-5,5,100)
y=np.linspace(-10,10,100)
x,y=np.meshgrid(x,y)


#function implementation
fun=0.26*(x**2+y**2)-0.48*x*y

#setting figure environment and drawing
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection="3d")
ax.plot_surface(x,y,fun,cmap=cm.nipy_spectral)


#setting scales
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('FUN')

plt.show()
```
