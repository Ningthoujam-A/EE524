```python
#QUESTION 1 

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 

fig = plt.figure()

ax1 = fig.add_subplot(121, projection = '3d')
X1 = np.linspace(-6, 6, 200)     
X2 = np.linspace(-6, 6, 200)     
X1, X2 = np.meshgrid(X1, X2)
Z1 = (X1**2 - 10 * np.cos(2 * np.pi * X1)) + (X2**2 - 10 * np.cos(2 * np.pi * X2)) + 20

ax1.plot_surface(X1, X2, Z1, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.08)

ax1.set_title('3D PLOT FOR RASTRIGIN ')

ax2 = fig.add_subplot(122)
X1 = np.linspace(-6, 6, 100)
Z2 = (X1**2 - 10 * np.cos(2 * np.pi * X1)) + 10 
ax2.plot(X1, Z2)
ax2.set_title('2D PLOT FOR RASTRIGIN FUNCTION')
plt.show()
plt.tight_layout()
```

```python
#QUESTION 2


import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 

fig = plt.figure()

ax1 = fig.add_subplot(121, projection = '3d')
X1 = np.linspace(-30, 30, 200)     
X2 = np.linspace(-30, 30, 200)     
X1, X2 = np.meshgrid(X1, X2)


b = 0.2
c = 2 * np.pi
sterm = -20 * np.exp(-b * np.sqrt(X1*X1 + X2*X2) / 2)
cterm = -np.exp((np.cos(c*X1) + np.cos(c*X2)) / 2)
Z = 0.2 + np.exp(1) + sterm + cterm

ax1.plot_surface(X1, X2, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral)

ax1.set_title('3D PLOT FOR ACKLEY ')

ax2 = fig.add_subplot(122)
X1 = np.linspace(-6, 6, 100)
c = 2 * np.pi
sterm = -20 * np.exp(-b * np.sqrt(X1*X1)) 
cterm =  -np.exp(np.cos(c*X1))

Z2 = 20 + np.exp(1) + sterm + cterm 
ax2.plot(X1, Z2)
ax2.set_title('2D PLOT FOR ACKLEY FUNCTION')
plt.show()



plt.tight_layout()
```

```python
#QUESTION 3

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 

fig = plt.figure()
ax = fig.gca(projection='3d')
X1 = np.linspace(-3, 3, 200)     
X2 = np.linspace(-3, 3, 200)     
X1, X2 = np.meshgrid(X1, X2)
Z = X1**2 + X2**2

ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,cmap=cm.coolwarm, linewidth=0.08)
ax.set_title('Sphere Function')
plt.show()
```

```python
#QUESTION 4

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X1 = np.linspace(-0.25,0.25, 100)     
X2 = np.linspace(-0.25, 0.25, 100)     
X1, X2 = np.meshgrid(X1, X2)

Z =100*((X1**2 - X2))**2  + (1-X1)**2 

fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,cmap=cm.hot, linewidth=0.08)
ax.set_title('Rosenbrock Function')
plt.show()
```

```python
#QUESTION 5

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-5,5, 200)     
Y = np.linspace(-5, 5, 200)     
X, Y = np.meshgrid(X, Y)

Z =  np.sin(X + Y) + (X-Y)**2 - 1.5*X + 2.5*Y + 1

fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=cm.rainbow, linewidth=0.08)
ax.set_title('McCormick Function')
plt.show()
```

```python
#QUESTION 6

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-4,4, 100)     
Y = np.linspace(-4, 4, 100)     
X, Y = np.meshgrid(X, Y)

Z =  0.26*(X**2 + Y**2) - 0.48*X*Y

fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral)
ax.set_title('Matyas Plot')
plt.show()
```
