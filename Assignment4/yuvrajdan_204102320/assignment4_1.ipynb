```python
import numpy as np              #importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy.stats import norm 
```

```python
x=np.linspace(-10,10,100)     # divide x into 100 parts even space
y=norm.pdf(x,0,1)            #we take probability density function of normal distribution
y1=norm.pdf(x,3,2)
```

```python
plt.plot(x,y)
plt.plot(x,y1)                # plotting the pdf of normal gaussian
plt.xlabel('value of x')
plt.ylabel('pdf value')
plt.title('gaussian pdf')
plt.legend(["blue (0,1)" ,"orange(3,2)"])
```

```python
mean =[0,0]                  #mean matrix
cov = [[1,0.6],[0.6,2]]      # diagonal covariance, points lie on x or y-axis
X=np.linspace(-10,10,100)
Y=np.linspace(-10,10,100)
X,Y=np.meshgrid(X,Y)
pos=np.dstack([X,Y])         #we convert stack 2D arrays into a single 3D array for processing
```

```python
from scipy.stats import multivariate_normal
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

F=multivariate_normal(mean,cov)             #multivariate normal distribution & density finction
Z=F.pdf(pos)                                 
```

```python
fig = plt.figure()                            #surface plot
ax = fig.gca(projection='3d') 
ax.set_title('gaussian surface plot')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=cm.nipy_spectral, linewidth=0.4,antialiased=True)    

plt.show()
```

```python
cp=plt.contourf(X,Y,Z )       #plot contour plot 
plt.colorbar(cp)
plt.xlabel('value of x')
plt.ylabel(' value of y')
plt.title('contour plot')
plt.show()
```
