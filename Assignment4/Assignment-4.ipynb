##### Q.1. (a). Let us consider x to be a random variable, where x belong to[-10,10]. Gen-erate 100 samples of x and plot the pdf of x for x~ N(0; 1) and x~N(3; 2).Comment on the difference in plot. If there is a difference in magnitude, comment why?
#### (b). Consider the two-dimensional normal distribution p(x|w) N(mue; sigma)where mue =[0,0]T and sigma =[[1 .6]:[.6 2]].Let x be a random vector denoted by x =[x1 x2], where xi=[-10,10]. Plot the surface plot and counter plot of the pdf?

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import axes3d
%matplotlib notebook
import math 
import random

#gerating sequence for x axis
x=np.linspace(-10,10,100)

#genrating first normal RV
mue_1=0
sig_1=1
var_1=sig_1**2
Y_1=(1/(2*math.pi*var_1)**.5)*((math.e)**-(((x-mue_1)*1/sig_1)**2))
plt.scatter(x,Y_1)

##genrating second normal RV
mue_2=3
sig_2=2
var_2=(sig_2)**2
Y_2=(1/(2*math.pi*var_2)**.5)*((math.e)**-(((x-mue_2)*1/sig_2)**2))
plt.scatter(x,Y_2)


#REASONING FOR MAGNITUDE DIFFERENCE
#magnitude is being governed by the variance alone whihc is diffenent for both of them
max_mag_1=1/(2*math.pi*var_1)**.5
print("maximum magnitudeof 1st pdf is",max_mag_1)
max_mag_2=1/(2*math.pi*var_2)**.5
print("maximum magnitudeof 2nd pdf is",max_mag_2)
diff_mag=max_mag_1-max_mag_2
print("difference in magnitude",diff_mag)
```

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

#reading a CSV file
attributes=["sepal_length","sepal_width","petal_length","petal_width","species"]
iris=pd.read_csv('iris.data',names=attributes)
print(iris.shape)
x=iris["species"].value_counts()
print(x)



#plotting unsorted 

iris.plot(kind='scatter',x="sepal_length", y="sepal_width")
plt.legend("species")
plt.show()
```

```python
#sorting the dataset
iris_sort=iris.sort_values("species",axis=0)
iris_sort
```

```python
#preparing training dataset
x=iris_sort.iloc[:,:-1]
print(x)
y=iris_sort.iloc[:,-1]
print(y)
```

```python
X=iris_sort.drop(columns=["species"])
Y=iris_sort["species"]
#spliting the  data as per the ratioa given
train_x,test_x=train_test_split(X,test_size=0.3)
train_y,test_y=train_test_split(Y,test_size=0.3)

print(train_x)
print(train_y)
print(test_x)
print(test_y)
```

```python
#model training
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(train_x,train_y)
```

```python
print("Accuracy:",model.score(test_x,test_y)*100)
```
