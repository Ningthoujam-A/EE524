```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
def class_separation(data):
    #SEPARATION OF CLASS OF THE DATA PROVIDED
    separated=dict()
    for i in range (len(data)):
        row=data[i]
        class_row=row[-1]
        if (class_row not in separated):
            separated[class_row]=list()
        separated[class_row].append(row)
    return(separated)
def SVM(X,y,percentage):
    X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=percentage,random_state=42)
    print("\n\nTraining data without class is\n",X_train,"\n\nTesting data is\n",X_test,"\n\nClass of training data is\n",y_train,"\n\nClass of testing data is\n",y_test)
    clf=svm.SVC(kernel='linear')
    clf.fit(X,y)
    y_pred=clf.predict(X_test)
    print("\n Confusion matrix for total of ",len(X_test)," test samples which is ",percentage*100,"% of test samples is shown below\n",confusion_matrix(y_test,y_pred))
    print("\n The precision report is shown below\n",classification_report(y_test,y_pred))
def nishan():
    data=[]
    #csvdata=[]
    keys=[]
    test_data=[]
    y=[]
    train_data=dict()
    file=open("iris.data")
    for line in file:
        data1=file.readline()
        data2=data1.rstrip().split(",")
        for i in range(len(data2)-1):
            data2[i]=float(data2[i])
        data.append(data2)
    file.close()
    post_separate=class_separation(data)
    X=[]
    for key in post_separate:
        for j in range(len(post_separate[key])):
            new_lst=[]
            for k in range(len(post_separate[key][0])):
                if k==len(post_separate[key][0])-1:
                    y.append(post_separate[key][j][k])
                else:
                    new_lst.append(post_separate[key][j][k])
            X.append(new_lst)
    SVM(X,y,0.75)
    SVM(X,y,0.8)
    
nishan()
```
