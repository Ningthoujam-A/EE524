```
from matplotlib import pyplot as plt
import numpy as np
```

```
# Minimum of f(x) = x^2+3x+2 using gradient descent

# Function for giving value of the derivative for any value of x
def derivative_of_fx(x):
  return (2*x)+3

iteration= 200 # Number of iteration
precision = 0.0001 # To stop the loop when it's close to the minima
i = 0 
x = 0 # Intial value of x
lr = 0.01 # Learning rate
step_size = 1 # Previous step size

# Finding x at which minima occurs
while i<iteration and precision < step_size:
  prev_x = x
  x = x - (lr*derivative_of_fx(prev_x))
  step_size = abs(x-prev_x)
  i += 1

print("The local minima occurs at "+str(x))
print("The minima of the function is "+str((x**2)+(3*x)+2))
```

```
# Minimum of f(x1,x2) = x1^2+ x2^2 using gradient descent

# Function for giving value of the derivative for any value of x
def derivative_of_fx(x):
  return (2*x)

iteration= 500 # Number of iteration
precision = 0.0001 # To stop the loop when it's close to the minima
i = 0 
x1 = 5 # Intial value of x1
x2 = 5 # Intial value of x2
lr = 0.01 # Learning rate
step_size = 1 # Previous step size

# Finding x at which minima occurs
while i<iteration and step_size > precision:
  prev_x1 = x1
  prev_x2 = x2
  x1 = x1 - (lr*derivative_of_fx(prev_x1))
  x2 = x2 - (lr*derivative_of_fx(prev_x2))
  step_size = min(abs(x1-prev_x1),abs(x2-prev_x2))
  i += 1

print("The local minima occurs at "+str(x1)+" "+str(x2))
print("The minima of the function is "+str((x1**2)+(x2**2)))
```

```
# Implementation of AND using perceptron

x=[[0,0],[0,1],[1,0],[1,1]] # The inputs
y=[0,0,0,1] # The desired output
lr=0.2 # Learning rate
theta=0.5 # threshold
weights=[0.5,0.5,0.5]
i=0
count=0
max_iteration = 10000

# Condition for iteration
while(count<4 and i< max_iteration):
    # Calculating the predicted output
    yc= weights[0]+(weights[1]*x[i%4][0])+(weights[2]*x[i%4][1])
    if (yc>theta):
        yc=1
    else:
        yc=0
    # Checking if the desired and predicted outputs are equal
    if(yc != y[i%4]):
        count=0
        # Updating weights according to the condition
        if(yc > y[i%4]):
            weights[0]=weights[0]-lr*1
            weights[1]=weights[1]-lr*x[i%4][0]
            weights[2]=weights[2]-lr*x[i%4][1]
        else:
            weights[0]=weights[0]+lr*1
            weights[1]=weights[1]+lr*x[i%4][0]
            weights[2]=weights[2]+lr*x[i%4][1]
        print("Iteration ",i," :",weights)
    else:
        count+=1
    i+=1

# Final weights
print(weights)

x = np.array(x)
# Plotting the input points
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')

# Calculating slope
xintr=(theta-weights[0])/weights[1]
yintr=(theta-weights[0])/weights[2]
slope=-yintr/xintr

# Plotting the decision boundary
xx=plt.xlim()
yeq = []
for x in xx:
    yeq.append(yintr+slope*x)
plt.plot(xx,yeq)
plt.show()
```
