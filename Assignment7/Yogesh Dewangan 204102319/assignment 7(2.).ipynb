# Minimize the function using the gradient descent algorithm.
# 3D function
#f(x1, x2) = x1^2+x2^2


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
```

```python
#Defining the function

def F(x):
    return x[0]**2 + x[1]**2
```

```python
#Defining the gradient of the function

def grad(x):
    g = [0]*2
    g[0]= 2*x[0]
    g[1]= 2*x[1]
    return g
```

```python
#Defining the gradient decent algorithm

def descent(x_new,x_prev,lr):
    #print(x_prev)
    #print(F(x_prev))
    a = []                    #creating the array with values of X1
    b = []                    #creating the array with values of X2
    c = []                    #creating the array with values of F(x)
    while True:
        x_prev = x_new
        x_0 = x_prev[0] - lr*grad(x_prev)[0]
        x_1 = x_prev[1] - lr*grad(x_prev)[1]
        x_new = [x_0,x_1]
        a.append(x_new[0])
        b.append(x_new[1])
        u = F(x_new)
        c.append(u)
        
        if (x_new[0]-x_prev[0])**2 + (x_new[1]-x_prev[1])**2 < pow(10,-7):
            return(u,a,b,c)
            break
```

```python
#Calling the fuction and finding the minimum value

f,A,B,C= descent ([5,5],[5,5], 0.05) 
print("The value of Minima of given function is ====>\n",f)
print("the value of x1 at which minimum occurs is===>\n",A[-1])
print("the value of x1 at which minimum occurs is===>\n",B[-1])
```

```python
#Plotiing the Traced values 

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Function value')
ax.set_title('Poit traced')
ax.scatter3D(A, B, C, color = "red")
plt.show()

```

```python
#plotting the traced value and the given function plot together

fig = plt.figure(1)
ax = fig.gca(projection='3d')
x1 = np.linspace(-5,5,3000)
x2 = np.linspace(-5,5,3000)
x1, x2 = np.meshgrid(x1, x2)
f_xy = (x1**2)+(x2**2)
ax.plot_surface(x1,x2,f_xy,cmap=cm.plasma) 
ax.scatter3D(A, B, C, color = "black")
ax.set_title('Value of function at each iteration')
ax.set_xlabel('X1----->')
ax.set_ylabel('X2------>')
ax.set_zlabel('F(x)------>')
plt.show()
```
