(Q.NO. 1)       GRADIENT DESCENT ALGORITHM for minimizing 2D function f(x) = x** 2 + 3* x +2 , and find its minima.

```python
import numpy as np
import matplotlib.pyplot as plt


learn_rate = 0.6
precision = 0.000001                ## to decide when the algoritm comes to halt
cur_x = 10                               # Initial point at x = 10
count = 0                                # Iterations count
maxi_iterate = 500                 ## total no of iterations 
previous_stepsize = 1

x = np.linspace(-10,10,100)
fx = x**2 + 3*x + 2                     # Defining fx
del_fx =lambda x : 2*x + 3             #  Differenciating fx

plt.plot(x,fx,color='b')

while previous_stepsize > precision and count < maxi_iterate:
    previous_x = cur_x                                             ## storing the current value of x in the previous_x
    cur_x = cur_x - (learn_rate)*del_fx(previous_x)            ## gradient descent
    previous_stepsize = abs(cur_x - previous_x )               ## incremental change in x 
    count = count + 1      
    
    
    fx1 = cur_x**2 + 3*cur_x + 2 
    
    plt.scatter(cur_x,fx1)
    plt.xlabel('range of x')
    plt.ylabel('diff(x)')
    plt.title('gradient descent of x**2 + 3*x + 2')
    
    
    
print ( "The minima of this function is at x = ", cur_x)    
```
