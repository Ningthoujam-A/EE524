## 3.) Boolean Function Learning

```python
# defining class perceptron for operation
class Perceptron(object):
    # defining parameters and rates
    def __init__(self, learning_rate=0.01, threshold=100, random_state=1):
        self.learning_rate = learning_rate
        self.threshold = threshold
        self.random_state = random_state
        
    def net_input(self, X):
        # z = w*x+b
        # forward propagation
        z = np.dot(X, self.weights[1:]) + self.weights[0]
        return z
    
    def predict(self, X):
        # step function
        return np.where(self.net_input(X) >= 0, 1, -1)
    
    def fit(self, X, y):
        rand = np.random.RandomState(self.random_state)
        # randomly selected weight
        self.weights = rand.normal(loc=0.0, scale=0.01, size= 1 +  X.shape[1])
        self.errors_ = []
        for _ in range(self.threshold):
            errors = 0
            for x, target in zip(X, y):
                update = self.learning_rate * (target - self.predict(x))
                # updating weights
                self.weights[1:] += update * x 
                # updating bias 
                self.weights[0] += update
                errors += int(update != 0.0)
                self.errors_.append(errors)
            return self 
```

## 3.1) y = f(x1, x2) = x1.x2

```python
# logical AND operation
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[0],[0],[1]])
plt.plot(X,y)
plt.show()
per = Perceptron(learning_rate=0.1, threshold=100, random_state=1)
per.fit(X, y)
plt.plot(range(1, len(per.errors_) + 1), per.errors_, marker='*')
plt.xlabel('iteration')
plt.ylabel('Number of updates')
plt.title("Boundary line ")
plt.show()
```
